```python
import random as rd
import matplotlib.pyplot as plt
x = range(2,26,2)
y = [rd.randint(15,30) for i in x]

plt.figure(figsize=(20,8),dpi=80)

plt.xlabel('time')
plt.ylabel('temperature')

x_label = ['{}:00'.format(i) for i in x]
y_label = ['{}°C'.format(i) for i in range(min(y),max(y)+1)]

plt.xticks(x,x_label,rotation=45)
plt.yticks(range(min(y),max(y)+1),y_label)

plt.plot(x,y)
plt.show()
```

```python
# 散点图
from matplotlib import pyplot as plt
from matplotlib import font_manager
import random

# 现实中文字体
plt.rcParams['font.family']=['Arial Unicode MS']
plt.rcParams['font.size']=16
plt.rcParams['axes.unicode_minus']=False

x = range(1,32)
y = [random.randint(6,30) for i in x]

plt.figure(figsize=(20,8),dpi=100)
plt.scatter(x,y,label='3月份')
# my_font  = font_manager.FontProperties(fname='./Microsoft YaHei.ttf',size=10)

_xticks_labels = ['3月{}日'.format(i) for i in x]
plt.xticks(x[::3],_xticks_labels[::3],rotation=45)
plt.xlabel('日期')
plt.ylabel('温度')

plt.legend()
plt.show()

```

```python
# 条形图
from matplotlib import pyplot as plt

# 现实中文字体
plt.rcParams['font.family']=['Arial Unicode MS']
plt.rcParams['font.size']=16
plt.rcParams['axes.unicode_minus']=False

x = ['流浪地球','疯狂外星人','飞驰人生','大黄蜂','熊出没','新喜剧之王']
y = ['38.13','19.85','14.89','11.36','6.47','5.93']

y= [float(i) for i in y]

plt.figure(figsize=(20,8),dpi=80)

rects = plt.bar(range(len(x)),y,width=0.5,color=['r','g','b','r','g','b'])

plt.xticks(range(len(x)),x)
plt.yticks(range(0,41,5),range(0,41,5))

for rect in rects:
    height = rect.get_height()
    plt.text(rect.get_x()+rect.get_width()/2,height+0.3,str(height),ha='center')

plt.show()



```

```python
# 横向条形图


# 条形图
from matplotlib import pyplot as plt

# 现实中文字体
plt.rcParams['font.family']=['Arial Unicode MS']
plt.rcParams['font.size']=16
plt.rcParams['axes.unicode_minus']=False

x = ['流浪地球','疯狂外星人','飞驰人生','大黄蜂','熊出没','新喜剧之王']
y = ['38.13','19.85','14.89','11.36','6.47','5.93']

y= [float(i) for i in y]

plt.figure(figsize=(20,8),dpi=80)

rects = plt.barh(range(len(x)),y,height=0.5,color=['r','g','b','r','g','b'])

plt.yticks(range(len(x)),x,rotation=45)
# plt.xticks(range(0,41,5),range(0,41,5))

for rect in rects:
    width = rect.get_width()
    plt.text(width,rect.get_y()+0.5/2,str(width),va='center')

plt.show()


```

```python
# 并列和罗列条形图

import matplotlib.pyplot as plt
import numpy as np
import random

index = np.arange(4)

BJ = [random.randint(10,20) for i in range(1,5)]
SH = [random.randint(10,30) for i in range(1,5)]

print(BJ)
print(SH)

plt.bar(index,BJ,width=0.3)
plt.bar(index+0.3,SH,width=0.3,color='g')

plt.xticks(index+0.3/2,index)

plt.show()

```

```python
from matplotlib import pyplot as plt
import random
time = [random.randint(80,180) for i in range(1,50)]
# print(time)

plt.figure(figsize=(20,8),dpi=100)

distance=5
group_num = int((max(time) - min(time))/distance)
plt.hist(time,bins=group_num,density=0)
 

#修改x刻度显示

plt.xticks(range(min(time),max(time)+2)[::5],rotation=45)

# 添加网格

plt.grid(linestyle='--',alpha=0.5)

plt.xlabel('电影时长')
plt.ylabel('电影的数据量')


plt.show()

```
